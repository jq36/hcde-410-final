# Coding

```python
import pandas as pd
df = pd.read_csv('./googleplaystore.csv')
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '')
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce').fillna(0).astype(int)

def convert_size(size):
    if pd.isnull(size):
        return None
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    else:
        return None

df['Size'] = df['Size'].replace('Varies with device', None)
df['Size'] = df['Size'].apply(convert_size)

df = df.dropna(subset=['Size', 'Rating'])

features = df[['Category', 'Reviews', 'Size', 'Installs', 'Type', 'Price', 'Content Rating']]
target = df['Rating']

features = pd.get_dummies(features)

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.4, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

feature_importances = pd.Series(model.feature_importances_, index=features.columns)
print(feature_importances.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_feature_importances(feature_importances, top_n=10):
    top_features = feature_importances.sort_values(ascending=False).head(top_n)

    sns.set(style="whitegrid")
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_features, y=top_features.index, palette="viridis")
    
    plt.title("Top Feature Importances", fontsize=16)
    plt.xlabel("Importance", fontsize=12)
    plt.ylabel("Features", fontsize=12)
    
    plt.tight_layout()
    plt.show()

plot_feature_importances(feature_importances)
plt.savefig('./plot_feature_importances.jpg', format='jpg', dpi=300)
```

## 6.4 Distribution of User Ratings

The following graph depicts the distribution of user ratings across all apps. It provides an overview of user satisfaction levels and helps identify patterns or anomalies in the dataset.

**Figure 2: Distribution of User Ratings**  


```python
plt.figure(figsize=(10, 6))
colors = {'FAMILY': 'skyblue', 'GAME': 'orange', 'TOOLS': 'green', 'MEDICAL': 'red'}

for category, color in colors.items():
    subset = df[df['Category'] == category]
    sns.histplot(subset['Rating'], bins=10, kde=True, color=color, label=category, alpha=0.5)

plt.title("Distribution of App Ratings by Category", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Number", fontsize=12)
plt.legend(title='Category')

plt.tight_layout()
plt.show()
plt.savefig('./distribution_of_app_rating_bycategory.jpg', format='jpg', dpi=300)
```
